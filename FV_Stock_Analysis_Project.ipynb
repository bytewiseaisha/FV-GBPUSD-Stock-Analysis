```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
dataset= pd.read_csv("C:/Users/AISHA/Desktop/Datasets/GBPUSD=X_dataset.csv")
```

```python
dataset
```

SCATTER PLOTS

```python
# Plot using a scatter plot matrix or pairplot

sns.pairplot(dataset)
plt.show()
```

```python
# Using several scatter plots to visualise for better view

columns = dataset.columns

for i in range(len(columns)):
    for j in range(i+1, len(columns)):
        # Create a new figure for each scatter plot
        plt.figure()
        plt.scatter(dataset[columns[i]], dataset[columns[j]])
        plt.xlabel(columns[i])
        plt.ylabel(columns[j])
        plt.title(f'{columns[i]} vs {columns[j]}')

        
        plt.close() # remove this line to see figures

        
plt.show()
```

```python
# From the scatter plots above, there are outliers
```

LINEAR MODEL FOR TRADE OPEN AND TRADE CLOSE

```python
# Specifying variables

x_1 = dataset[["Open"]]
y_1 = dataset[["Close"]]
```

```python
# Train-test split

x_train_1, x_test_1, y_train_1, y_test_1 = train_test_split(x_1,y_1, test_size=0.2, random_state=42)
```

```python
# Creating and training model

model_1 =  LinearRegression()
model_1.fit(x_train_1, y_train_1)
```

```python
# Making predictions on test data

y_pred_1 = model_1.predict(x_test_1)
y_pred_1
```

```python
# Evaluating model

mse_1 = mean_squared_error (y_test_1, y_pred_1)
print (f" The mean squared error of the linear model for values of trade open and trade close is:  {mse_1}")
```

```python
# Slope
m_1 = model_1.coef_
print (f" The slope of the linear model for values of trade open and trade close is: {m_1}")
```

```python
# Intercept
c_1 = model_1.intercept_
print (f" The intercept of the linear model for values of trade open and trade close is: {c_1}")
```

```python
# r_squared
r_squared_1 = r2_score(y_test_1, y_pred_1)
print (f" The r-squared value of the linear model for values of trade open and trade close is: {r_squared_1}")
```

```python
# Correlation
correlation_matrix_1 = dataset[["Open", "Close"]].corr()
correlation_matrix_1
```
